```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ageList = [10,20,30,40,50,60,70,80,90]
weightList = [20,70,75,70,80,120,67,117,100]
```

```python
np_ageList = np.array(ageList)
np_weightList= np.array(weightList)
```

```python
plt.plot(np_ageList, np_weightList, "r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Weight graph's according to Age")
# Biz baştaki inline olayını yazmasak plt.show() yapmalıydık
```

```python
numpyArray1 = np.linspace(0, 30, 20)
numpyArray2 = numpyArray1 ** 2
```

```python
plt.plot(numpyArray1, numpyArray2, "b*-")
```

```python
plt.subplot(1, 2, 1)# 1 sıra olacak, 2 kolon olacak, şuan 1.grafiği çiziyorum
plt.plot(numpyArray1, numpyArray2, "r--")
plt.subplot(1, 2, 2)# 1 sıra olacak, 2 kolon olacak, şuan 2.grafiği çiziyorum
plt.plot(numpyArray1, numpyArray2, "g*-")
```

```python
myFigure = plt.figure(dpi=100)
figureAxes = myFigure.add_axes([0.4, 0.4, 0.7, 0.9])
figureAxes.plot(numpyArray1, numpyArray2, "g")
figureAxes.set_xlabel("X Axes")
figureAxes.set_ylabel("Y Axes")
figureAxes.set_title("Graph Title")
# Şimdiye kadar plt kütüphanesinin sunduğu 
# fonskiyonlardan yararlanarak grafik oluşturduk
# bu örnekte ise yeni bir figur(canvas) oluşrutup
# onun içini özelleştiriyoruz.
```

```python
newFigure = plt.figure(dpi=100)

new_figureAxes1 = newFigure.add_axes([0.1, 0.1, 0.7, 0.7])
new_figureAxes2 = newFigure.add_axes([0.3, 0.3, 0.2, 0.2])

new_figureAxes1.plot(numpyArray1, numpyArray2, "g")
new_figureAxes1.set_xlabel("X Axes")
new_figureAxes1.set_ylabel("Y Axes")
new_figureAxes1.set_title("Big Graph Title")

new_figureAxes2.plot(numpyArray1, numpyArray2, "r")
new_figureAxes2.set_xlabel("X Axes")
new_figureAxes2.set_ylabel("Y Axes")
new_figureAxes2.set_title("Small Graph Title")
```

```python
(Figure_new, Axes_new) = plt.subplots(nrows=1, ncols=2)
# Yukardaki işlemde plt.subplots(ncolumn=2) 
# sayesinde bir dizi döndürülüyor.

for axes in Axes_new:
    axes.plot(numpyArray1, numpyArray2, "g")
    axes.set_xlabel("X Axes")
plt.tight_layout() # Daha düzgün görünüm
```

```python
coolFigure = plt.figure(dpi=500)

coolAxes = coolFigure.add_axes([0.1, 0.1, 0.9, 0.9])

coolAxes.plot(numpyArray1, numpyArray2, label= "numpyArray1 **2")
coolAxes.plot(numpyArray1, numpyArray1 **3, label= "numpyArray1 **3")

coolAxes.legend(loc = 2)
```

```python
# coolFigure.savefig("coolFigure2.png", dpi=90) Kaydetmek için
# Herhangi bir axes için plot() çağırdığımızda içinde 
# plot(color= "#C9F89E", alpha = 0.3) ile renk ve transparanslık sağlayabiliriz.
# aynı şekilde "linewidth" ile çizgilerin kalınlığını da değiştirebiliriz
```

```python
(styleFigure, styleAxes) = plt.subplots()

styleAxes.plot(numpyArray1, numpyArray1 + 2, color = "#000000", linestyle = "--", marker = "o", markersize = 4, markerfacecolor = "r")
```
