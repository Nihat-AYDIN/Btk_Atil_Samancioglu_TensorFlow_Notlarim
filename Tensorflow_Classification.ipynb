```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 15)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Activation, Dropout
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units = 30, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))

model.add(Dense(units = 1, activation = "sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
# Bu denemede normalde vermem gerekende
# fazla bir epoch değeri verirsem overfitting
# olduğunu görüyoruz.

model.fit(x = x_train, y = y_train, epochs = 700, validation_data = (x_test, y_test), verbose = 1) 
```

```python
modeLoss = pd.DataFrame(model.history.history)
```

```python
modeLoss.plot()
```

```python
# Yukarda da görüldüğü gibi modelimizin sonucu
# olan loss değeri(mavi çizgi) olması gerekenden
# daha fazla aşağı inmiş, yani bizim ona verdiğimiz
# train değerlerini ezberlemiş diyebiliriz. Bunun 
# sebebi bizim çok fazla "epoch" değeri vermemiz.
```

## Early Stopping

```python
model = Sequential()

model.add(Dense(units = 30, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))

model.add(Dense(units = 1, activation = "sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
# Yukarıda anlatılan overfitting durumunu engellemek
# için earlyStopping kullanılır. Aşağıda monitor etmesi
# için "val_loss" verilmiştir. Mode değeri monitor ettiği
# değerin nasıl olması istendiğine göre değişir. Mesela
# eğer biz accuracy değerini monitor etseydik max olmasını
# isterdik. Patience ise kaç epoch'ta bir check edilmesi
# gerektiğini belirtir.
earlyStopping = EarlyStopping(monitor = "val_loss", mode = "min", verbose = 1, patience = 25)
```

```python
model.fit(x = x_train, y = y_train, epochs = 700,
          validation_data = (x_test, y_test), verbose = 1, callbacks = [earlyStopping]) 
```

```python
modeLossTwo = pd.DataFrame(model.history.history) 
```

```python
modeLossTwo.plot()
```

## Dropout

```python
model = Sequential()

model.add(Dense(units = 30, activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units = 15, activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units = 15, activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units = 1, activation = "sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
model.fit(x = x_train, y = y_train, epochs = 700,
          validation_data = (x_test, y_test), verbose = 1, callbacks = [earlyStopping]) 
```

```python
modeLossThree = pd.DataFrame(model.history.history) 
```

```python
modeLossThree.plot()
```

```python
ourPredicts = (model.predict(x_test) > 0.5).astype("int32")
```

```python
ourPredicts
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, ourPredicts))
```

```python
# Precision, sınıflandırıcınızın "pozitif" olarak
# sınıflandırdığı örneklerin kaçının gerçekten 
# pozitif olduğunu gösterir.

# Recall, gerçekten pozitif olan örneklerin kaçının
# sınıflandırıcınız tarafından doğru bir şekilde
# tespit edildiğini gösterir. 

# F1-score, precision ve recall metriklerinin 
# harmonik ortalamasını verir ve bir sınıflandırıcının 
# genel performansını özetler.
```

```python
from sklearn import metrics
```

```python
confusion_matrix = metrics.confusion_matrix(y_test, ourPredicts)
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
```

```python
cm_display.plot()
```
